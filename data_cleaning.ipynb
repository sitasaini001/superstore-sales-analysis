```python
import pandas as pd
```

```python
df=pd.read_csv("superstore_sales_dataset.csv")
```

```python
df
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"]  = pd.to_datetime(df["Ship Date"],  dayfirst=True)
```

```python
df
```

```python
data_types=df.dtypes
data_types
```

```python
print(df.isnull().sum())
```

```python
df["Postal Code"]=df["Postal Code"].fillna("Unknown")
print(df.isnull().sum())
```

```python
# Time features from Order Date
df["Year"]  = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Quarter"] = df["Order Date"].dt.quarter
df["Month Name"] = df["Order Date"].dt.strftime("%b")   # Jan, Feb ...
df["Day of Week"] = df["Order Date"].dt.day_name()
```

```python
# Fulfilment time (key derived metric)
df["Fulfilment Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# Order size (total sales per order)
df["Order Total"] = df.groupby("Order ID")["Sales"].transform("sum")

# Sales rank within category
df["Sales Rank"] = df.groupby("Category")["Sales"].rank(ascending=False, method="dense")
```

```python
df["Order Size"]=df.groupby('Order ID')['Sales'].transform('sum')
```

```python
df.to_csv("superstore_sales_dataset_clean.csv", index=False)
print(f"Saved: {df.shape[0]} rows, {df.shape[1]} columns")
```
